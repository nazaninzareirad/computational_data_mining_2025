1. spectral analysis of model weights

1.1 Load pretrained ResNet18

```python
import torch
import torchvision.models as models
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd

print("Loading pretrained ResNet18...")
model = models.resnet18(pretrained=True)
model.eval()
```

1.2 extract weight from middle, final layers

```python
# middle layer: layer3 (conv layer)
middle_layer_weights = model.layer3[0].conv1.weight.data.cpu().numpy()
# reshape to 2D for SVD
middle_weights_2d = middle_layer_weights.reshape(middle_layer_weights.shape[0], -1)

# final layer: fully connected
final_layer_weights = model.fc.weight.data.cpu().numpy()

print(f"Middle layer shape: {middle_weights_2d.shape}")
print(f"Final layer shape: {final_layer_weights.shape}")
```

1.3 singular value plotting

```python
def analyze_singular_values(weights, layer_name):
    # compute SVD
    U, s, Vt = svd(weights, full_matrices=False)

    # calculate variance explained
    variance = s ** 2
    variance_ratio = variance / variance.sum()
    cumulative_variance = np.cumsum(variance_ratio)

    # find number of components for 95% variance
    n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
    percentage_95 = (n_components_95 / len(s)) * 100

    # plot singular values
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # plot 1: Singular values
    axes[0].plot(s, 'b-', linewidth=2)
    axes[0].set_xlabel('Index')
    axes[0].set_ylabel('Singular Value')
    axes[0].set_title(f'{layer_name} - Singular Values')
    axes[0].grid(True, alpha=0.3)
    axes[0].set_yscale('log')

    # plot 2: Cumulative variance
    axes[1].plot(cumulative_variance, 'r-', linewidth=2)
    axes[1].axhline(y=0.95, color='g', linestyle='--', label='95% threshold')
    axes[1].axvline(x=n_components_95, color='orange', linestyle='--',
                    label=f'{n_components_95} components')
    axes[1].set_xlabel('Number of Components')
    axes[1].set_ylabel('Cumulative Variance Explained')
    axes[1].set_title(f'{layer_name} - Cumulative Variance')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # print analysis
    print(f"\n{layer_name} Analysis:")
    print(f"Total singular values: {len(s)}")
    print(f"Components for 95% variance: {n_components_95}")
    print(f"Percentage of components needed: {percentage_95:.2f}%")
    print(f"Top 5 singular values: {s[:5]}")

    return s, cumulative_variance, n_components_95

# analyze middle layer
print("\n" + "="*50)
s_middle, cum_var_middle, n95_middle = analyze_singular_values(
    middle_weights_2d, "Middle Layer (layer3.conv1)"
)

# analyze final layer
print("\n" + "="*50)
s_final, cum_var_final, n95_final = analyze_singular_values(
    final_layer_weights, "Final Layer (FC)"
)

# summary comparison
print("\n" + "="*50)
print("SUMMARY:")
print(f"Middle layer: {n95_middle}/{len(s_middle)} components ({(n95_middle/len(s_middle)*100):.2f}%) for 95% variance")
print(f"Final layer: {n95_final}/{len(s_final)} components ({(n95_final/len(s_final)*100):.2f}%) for 95% variance")
```

2. Practical Compression Implementation

2.1 Compress linear layer using rank-k SVD

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torch.utils.data import DataLoader
import numpy as np
import pandas as pd
from scipy.linalg import svd
import copy

def compress_linear_layer(layer, rank):
    W = layer.weight.data.cpu().numpy()
    U, s, Vt = svd(W, full_matrices=False)

    # truncate to rank k
    U_k, s_k, Vt_k = U[:, :rank], s[:rank], Vt[:rank, :]

    # create two layers: Vt_k and U_k @ diag(s_k)
    layer1 = nn.Linear(layer.in_features, rank, bias=False)
    layer1.weight.data = torch.from_numpy(Vt_k).float()

    layer2 = nn.Linear(rank, layer.out_features, bias=layer.bias is not None)
    layer2.weight.data = torch.from_numpy(U_k @ np.diag(s_k)).float()
    if layer.bias is not None:
        layer2.bias.data = layer.bias.data.clone()

    return nn.Sequential(layer1, layer2)


def compress_model(model, compression_ratio):
    compressed_model = copy.deepcopy(model)
    W = compressed_model.fc.weight.data.cpu().numpy()
    _, s, _ = svd(W, full_matrices=False)
    rank = max(1, int(len(s) * compression_ratio))
    compressed_model.fc = compress_linear_layer(compressed_model.fc, rank)
    return compressed_model


def count_parameters(model):
    return sum(p.numel() for p in model.parameters())


def evaluate_model(model, dataloader, device, max_batches=50):
    model.eval()
    correct = total = 0

    with torch.no_grad():
        for batch_idx, (images, labels) in enumerate(dataloader):
            if batch_idx >= max_batches:
                break
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    return 100 * correct / total


# setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}\n")

# load model
original_model = models.resnet18(pretrained=True).to(device)

# prepare validation data
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
```
