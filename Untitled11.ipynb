1. spectral analysis of model weights

1.1 Load pretrained ResNet18

```python
import torch
import torchvision.models as models
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd

print("Loading pretrained ResNet18...")
model = models.resnet18(pretrained=True)
model.eval()
```

1.2 extract weight from middle, final layers

```python
# middle layer: layer3 (conv layer)
middle_layer_weights = model.layer3[0].conv1.weight.data.cpu().numpy()
# reshape to 2D for SVD
middle_weights_2d = middle_layer_weights.reshape(middle_layer_weights.shape[0], -1)

# final layer: fully connected
final_layer_weights = model.fc.weight.data.cpu().numpy()

print(f"Middle layer shape: {middle_weights_2d.shape}")
print(f"Final layer shape: {final_layer_weights.shape}")
```
