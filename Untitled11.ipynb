1. spectral analysis of model weights

1.1 Load pretrained ResNet18

```python
import torch
import torchvision.models as models
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd

print("Loading pretrained ResNet18...")
model = models.resnet18(pretrained=True)
model.eval()
```

1.2 extract weight from middle, final layers

```python
# middle layer: layer3 (conv layer)
middle_layer_weights = model.layer3[0].conv1.weight.data.cpu().numpy()
# reshape to 2D for SVD
middle_weights_2d = middle_layer_weights.reshape(middle_layer_weights.shape[0], -1)

# final layer: fully connected
final_layer_weights = model.fc.weight.data.cpu().numpy()

print(f"Middle layer shape: {middle_weights_2d.shape}")
print(f"Final layer shape: {final_layer_weights.shape}")
```

1.3 singular value plotting

```python
def analyze_singular_values(weights, layer_name):
    # compute SVD
    U, s, Vt = svd(weights, full_matrices=False)

    # calculate variance explained
    variance = s ** 2
    variance_ratio = variance / variance.sum()
    cumulative_variance = np.cumsum(variance_ratio)

    # find number of components for 95% variance
    n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
    percentage_95 = (n_components_95 / len(s)) * 100

    # plot singular values
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # plot 1: Singular values
    axes[0].plot(s, 'b-', linewidth=2)
    axes[0].set_xlabel('Index')
    axes[0].set_ylabel('Singular Value')
    axes[0].set_title(f'{layer_name} - Singular Values')
    axes[0].grid(True, alpha=0.3)
    axes[0].set_yscale('log')

    # plot 2: Cumulative variance
    axes[1].plot(cumulative_variance, 'r-', linewidth=2)
    axes[1].axhline(y=0.95, color='g', linestyle='--', label='95% threshold')
    axes[1].axvline(x=n_components_95, color='orange', linestyle='--',
                    label=f'{n_components_95} components')
    axes[1].set_xlabel('Number of Components')
    axes[1].set_ylabel('Cumulative Variance Explained')
    axes[1].set_title(f'{layer_name} - Cumulative Variance')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # print analysis
    print(f"\n{layer_name} Analysis:")
    print(f"Total singular values: {len(s)}")
    print(f"Components for 95% variance: {n_components_95}")
    print(f"Percentage of components needed: {percentage_95:.2f}%")
    print(f"Top 5 singular values: {s[:5]}")

    return s, cumulative_variance, n_components_95

# analyze middle layer
print("\n" + "="*50)
s_middle, cum_var_middle, n95_middle = analyze_singular_values(
    middle_weights_2d, "Middle Layer (layer3.conv1)"
)

# analyze final layer
print("\n" + "="*50)
s_final, cum_var_final, n95_final = analyze_singular_values(
    final_layer_weights, "Final Layer (FC)"
)

# summary comparison
print("\n" + "="*50)
print("SUMMARY:")
print(f"Middle layer: {n95_middle}/{len(s_middle)} components ({(n95_middle/len(s_middle)*100):.2f}%) for 95% variance")
print(f"Final layer: {n95_final}/{len(s_final)} components ({(n95_final/len(s_final)*100):.2f}%) for 95% variance")
```

2. Practical Compression Implementation

2.1 Compress linear layer using rank-k SVD

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torch.utils.data import DataLoader
import numpy as np
import pandas as pd
from scipy.linalg import svd
import copy

def compress_linear_layer(layer, rank):
    W = layer.weight.data.cpu().numpy()
    U, s, Vt = svd(W, full_matrices=False)

    # truncate to rank k
    U_k, s_k, Vt_k = U[:, :rank], s[:rank], Vt[:rank, :]

    # create two layers: Vt_k and U_k @ diag(s_k)
    layer1 = nn.Linear(layer.in_features, rank, bias=False)
    layer1.weight.data = torch.from_numpy(Vt_k).float()

    layer2 = nn.Linear(rank, layer.out_features, bias=layer.bias is not None)
    layer2.weight.data = torch.from_numpy(U_k @ np.diag(s_k)).float()
    if layer.bias is not None:
        layer2.bias.data = layer.bias.data.clone()

    return nn.Sequential(layer1, layer2)


def compress_model(model, compression_ratio):
    compressed_model = copy.deepcopy(model)
    W = compressed_model.fc.weight.data.cpu().numpy()
    _, s, _ = svd(W, full_matrices=False)
    rank = max(1, int(len(s) * compression_ratio))
    compressed_model.fc = compress_linear_layer(compressed_model.fc, rank)
    return compressed_model


def count_parameters(model):
    return sum(p.numel() for p in model.parameters())


def evaluate_model(model, dataloader, device, max_batches=50):
    model.eval()
    correct = total = 0

    with torch.no_grad():
        for batch_idx, (images, labels) in enumerate(dataloader):
            if batch_idx >= max_batches:
                break
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    return 100 * correct / total


# setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}\n")

# load model
original_model = models.resnet18(pretrained=True).to(device)

# prepare validation data
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
```

2.2 Create and evaluate compressed models

```python
print("Creating compressed models...")
model_50 = compress_model(original_model, 0.50).to(device)
model_20 = compress_model(original_model, 0.20).to(device)

print("\nEvaluating models...")
acc_original = evaluate_model(original_model, val_loader, device)
acc_50 = evaluate_model(model_50, val_loader, device)
acc_20 = evaluate_model(model_20, val_loader, device)

params_original = count_parameters(original_model)
params_50 = count_parameters(model_50)
params_20 = count_parameters(model_20)
```

3. Accuracy Recovery and Speed Evaluation

3.1 fine-tune compressed model

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.linalg import svd
import copy

# Reuse compression functions from Part 2
def compress_linear_layer(layer, rank):
    W = layer.weight.data.cpu().numpy()
    U, s, Vt = svd(W, full_matrices=False)
    U_k, s_k, Vt_k = U[:, :rank], s[:rank], Vt[:rank, :]

    layer1 = nn.Linear(layer.in_features, rank, bias=False)
    layer1.weight.data = torch.from_numpy(Vt_k).float()

    layer2 = nn.Linear(rank, layer.out_features, bias=layer.bias is not None)
    layer2.weight.data = torch.from_numpy(U_k @ np.diag(s_k)).float()
    if layer.bias is not None:
        layer2.bias.data = layer.bias.data.clone()

    return nn.Sequential(layer1, layer2)


def compress_model(model, compression_ratio):
    compressed_model = copy.deepcopy(model)
    W = compressed_model.fc.weight.data.cpu().numpy()
    _, s, _ = svd(W, full_matrices=False)
    rank = max(1, int(len(s) * compression_ratio))
    compressed_model.fc = compress_linear_layer(compressed_model.fc, rank)
    return compressed_model


# Fine-tune compressed model
def fine_tune_model(model, train_loader, val_loader, device, epochs=3, lr=0.001):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    train_accs, val_accs = [], []

    for epoch in range(epochs):
        # Training
        model.train()
        correct = total = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_acc = 100 * correct / total
        train_accs.append(train_acc)

        # Validation
        model.eval()
        correct = total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        val_acc = 100 * correct / total
        val_accs.append(val_acc)

        print(f"Epoch {epoch+1}/{epochs} - Train Acc: {train_acc:.2f}%, Val Acc: {val_acc:.2f}%")

    return model, train_accs, val_accs
```

3.2 inference speed on cpu

```python
def measure_inference_speed(model, input_size=(1, 3, 224, 224), device='cpu', iterations=100):
    model = model.to(device)
    model.eval()

    dummy_input = torch.randn(input_size).to(device)

    # Warmup
    for _ in range(10):
        with torch.no_grad():
            _ = model(dummy_input)

    # Measure
    times = []
    with torch.no_grad():
        for _ in range(iterations):
            start = time.time()
            _ = model(dummy_input)
            times.append(time.time() - start)

    return np.mean(times) * 1000  # Convert to ms


# Setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}\n")

# Prepare data
transform_train = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

transform_val = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
val_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_val)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=128, shuffle=False, num_workers=2)

# Load pretrained model
original_model = models.resnet18(pretrained=True)
original_model.fc = nn.Linear(original_model.fc.in_features, 10)  # Adapt for CIFAR10

# Test multiple compression ratios
compression_ratios = [1.0, 0.8, 0.6, 0.5, 0.4, 0.3, 0.2]
results = {'ratio': [], 'accuracy_before': [], 'accuracy_after': [], 'speed_cpu': []}

print("="*70)
print("FINE-TUNING AND EVALUATION")
print("="*70)

for ratio in compression_ratios:
    print(f"\n{'='*70}")
    print(f"Compression Ratio: {ratio*100:.0f}% (keep {ratio*100:.0f}% of singular values)")
    print(f"{'='*70}")

    # Create compressed model
    if ratio == 1.0:
        model = copy.deepcopy(original_model)
    else:
        model = compress_model(original_model, ratio)

    # Evaluate before fine-tuning
    model.eval()
    model = model.to(device)
    correct = total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    acc_before = 100 * correct / total

    print(f"Accuracy before fine-tuning: {acc_before:.2f}%")

    # Fine-tune for 3 epochs
    print("Fine-tuning for 3 epochs...")
    model, train_accs, val_accs = fine_tune_model(model, train_loader, val_loader, device, epochs=3)
    acc_after = val_accs[-1]

    print(f"Accuracy after fine-tuning: {acc_after:.2f}%")
    print(f"Improvement: {acc_after - acc_before:.2f}%")

    # Measure CPU inference speed
    print("Measuring CPU inference speed...")
    speed = measure_inference_speed(model, device='cpu')
    print(f"Average inference time: {speed:.2f} ms")

    results['ratio'].append(ratio)
    results['accuracy_before'].append(acc_before)
    results['accuracy_after'].append(acc_after)
    results['speed_cpu'].append(speed)
```

```python
# Part 4: Analytical Questions

import torch
import torch.nn as nn
import torchvision.models as models
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
import time
import pandas as pd

# Load model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = models.resnet18(weights='DEFAULT')
model.eval()

# Question 4.1: Analyze latency impact of increased layers in different environments
print("="*70)
print("QUESTION 4.1: LAYER DECOMPOSITION LATENCY ANALYSIS")
print("="*70)

def create_decomposed_layer(in_features, out_features, rank):
    """Create rank-k decomposed layer"""
    layer1 = nn.Linear(in_features, rank, bias=False)
    layer2 = nn.Linear(rank, out_features, bias=True)
    return nn.Sequential(layer1, layer2)

def measure_latency(layer, input_size, device, iterations=1000):
    """Measure layer latency"""
    layer = layer.to(device)
    layer.eval()
    dummy_input = torch.randn(1, input_size).to(device)

    # Warmup
    for _ in range(50):
        with torch.no_grad():
            _ = layer(dummy_input)

    # Measure
    times = []
    with torch.no_grad():
        for _ in range(iterations):
            if device.type == 'cuda':
                torch.cuda.synchronize()
            start = time.time()
            _ = layer(dummy_input)
            if device.type == 'cuda':
                torch.cuda.synchronize()
            times.append(time.time() - start)

    return np.mean(times) * 1000  # ms

# Test FC layer with different ranks
in_feat, out_feat = 512, 1000
ranks = [100, 200, 300, 400, 500]

print(f"\nTesting FC layer: {in_feat} → {out_feat}")
print(f"Original parameters: {in_feat * out_feat:,}\n")

results_cpu = []
results_gpu = []

# Original layer
original_layer = nn.Linear(in_feat, out_feat)
lat_cpu_orig = measure_latency(original_layer, in_feat, torch.device('cpu'))
lat_gpu_orig = measure_latency(original_layer, in_feat, device) if device.type == 'cuda' else 0

print(f"Original layer:")
print(f"  CPU: {lat_cpu_orig:.4f} ms")
if device.type == 'cuda':
    print(f"  GPU: {lat_gpu_orig:.4f} ms")

# Decomposed layers
for rank in ranks:
    decomposed = create_decomposed_layer(in_feat, out_feat, rank)
    params = in_feat * rank + rank * out_feat
    compression = (1 - params / (in_feat * out_feat)) * 100

    lat_cpu = measure_latency(decomposed, in_feat, torch.device('cpu'))
    lat_gpu = measure_latency(decomposed, in_feat, device) if device.type == 'cuda' else 0

    overhead_cpu = ((lat_cpu - lat_cpu_orig) / lat_cpu_orig) * 100
    overhead_gpu = ((lat_gpu - lat_gpu_orig) / lat_gpu_orig) * 100 if device.type == 'cuda' else 0

    results_cpu.append({'rank': rank, 'latency': lat_cpu, 'overhead': overhead_cpu})
    results_gpu.append({'rank': rank, 'latency': lat_gpu, 'overhead': overhead_gpu})

    print(f"\nRank {rank} ({compression:.1f}% compression):")
    print(f"  Parameters: {params:,}")
    print(f"  CPU: {lat_cpu:.4f} ms (overhead: {overhead_cpu:+.1f}%)")
    if device.type == 'cuda':
        print(f"  GPU: {lat_gpu:.4f} ms (overhead: {overhead_gpu:+.1f}%)")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot CPU latency
axes[0].axhline(y=lat_cpu_orig, color='r', linestyle='--', label='Original', linewidth=2)
axes[0].plot(ranks, [r['latency'] for r in results_cpu], 'o-', linewidth=2, markersize=8)
axes[0].set_xlabel('Rank (k)')
axes[0].set_ylabel('Latency (ms)')
axes[0].set_title('CPU Inference Latency vs Rank')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot overhead
axes[1].plot(ranks, [r['overhead'] for r in results_cpu], 'o-', label='CPU', linewidth=2, markersize=8)
if device.type == 'cuda':
    axes[1].plot(ranks, [r['overhead'] for r in results_gpu], 's-', label='GPU', linewidth=2, markersize=8)
axes[1].axhline(y=0, color='r', linestyle='--', linewidth=1)
axes[1].set_xlabel('Rank (k)')
axes[1].set_ylabel('Latency Overhead (%)')
axes[1].set_title('Latency Overhead from Layer Decomposition')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("ANALYSIS:")
print("="*70)
print("1. CPU Environment:")
print("   - Decomposition adds overhead due to sequential matrix ops")
print("   - Lower ranks reduce parameters but increase layer calls")
print("   - Overhead ranges from ~{}% to ~{}%".format(
    min(r['overhead'] for r in results_cpu),
    max(r['overhead'] for r in results_cpu)))

if device.type == 'cuda':
    print("\n2. GPU Environment:")
    print("   - Overhead is typically lower due to parallel processing")
    print("   - Memory bandwidth becomes critical factor")
    print("   - Overhead ranges from ~{:.1f}% to ~{:.1f}%".format(
        min(r['overhead'] for r in results_gpu),
        max(r['overhead'] for r in results_gpu)))

print("\n3. Trade-off:")
print("   - Lower ranks: fewer parameters, faster memory, but more overhead")
print("   - Optimal rank balances compression and latency")

# Question 4.2: Using condition number to select compressible layers
print("\n" + "="*70)
print("QUESTION 4.2: CONDITION NUMBER FOR COMPRESSION")
print("="*70)

def compute_condition_number(weights):
    """Compute condition number from SVD"""
    _, s, _ = svd(weights, full_matrices=False)
    return s[0] / s[-1] if s[-1] > 1e-10 else np.inf

def analyze_layer_compressibility(weights, layer_name):
    """Analyze layer using condition number and singular values"""
    _, s, _ = svd(weights, full_matrices=False)

    # Condition number
    cond_num = compute_condition_number(weights)

    # Effective rank (95% variance)
    variance = s ** 2
    variance_ratio = variance / variance.sum()
    cumulative = np.cumsum(variance_ratio)
    effective_rank = np.argmax(cumulative >= 0.95) + 1

    # Singular value decay rate
    s_norm = s / s[0]
    decay_rate = np.mean(np.diff(np.log(s_norm[:min(20, len(s))])))

    return {
        'layer': layer_name,
        'shape': weights.shape,
        'condition_number': cond_num,
        'effective_rank': effective_rank,
        'rank_ratio': effective_rank / min(weights.shape),
        'decay_rate': decay_rate,
        'compressibility': 'High' if cond_num > 100 and effective_rank / min(weights.shape) < 0.5 else 'Medium' if cond_num > 10 else 'Low'
    }

# Analyze all linear/conv layers
layer_analysis = []

# FC layer
fc_weights = model.fc.weight.data.cpu().numpy()
layer_analysis.append(analyze_layer_compressibility(fc_weights, 'FC'))

# Conv layers (sample from each block)
conv_layers = [
    ('layer1[0].conv1', model.layer1[0].conv1),
    ('layer2[0].conv1', model.layer2[0].conv1),
    ('layer3[0].conv1', model.layer3[0].conv1),
    ('layer4[0].conv1', model.layer4[0].conv1),
]

for name, conv in conv_layers:
    w = conv.weight.data.cpu().numpy()
    w_2d = w.reshape(w.shape[0], -1)
    layer_analysis.append(analyze_layer_compressibility(w_2d, name))

# Create summary table
df = pd.DataFrame(layer_analysis)
df['condition_number'] = df['condition_number'].apply(lambda x: f"{x:.2e}")
df['rank_ratio'] = df['rank_ratio'].apply(lambda x: f"{x:.3f}")
df['decay_rate'] = df['decay_rate'].apply(lambda x: f"{x:.4f}")

print("\nLayer Compressibility Analysis:")
print(df.to_string(index=False))

print("\n" + "="*70)
print("CONDITION NUMBER INTERPRETATION:")
print("="*70)
print("""
1. Condition Number (κ = σ_max / σ_min):
   - Low (κ < 10): Well-conditioned, stable, less compressible
   - Medium (10 < κ < 100): Moderate compression possible
   - High (κ > 100): Ill-conditioned, highly compressible

2. Selection Strategy:
   - Target layers with high condition numbers
   - Low rank_ratio indicates redundancy
   - Fast decay_rate suggests compressibility

3. Practical Guidelines:
   - FC layers: Usually high κ, good compression candidates
   - Early conv layers: Lower κ, preserve for feature extraction
   - Late conv layers: Higher κ, can be compressed
   - Check effective rank < 50% of full rank for compression

4. Compression Priority:
""")

# Sort by compressibility
sorted_layers = sorted(layer_analysis,
                       key=lambda x: (x['compressibility'] == 'High',
                                     float(x['condition_number']),
                                     -x['rank_ratio']),
                       reverse=True)

for i, layer in enumerate(sorted_layers[:3], 1):
    print(f"   {i}. {layer['layer']}: κ={layer['condition_number']:.2e}, "
          f"rank_ratio={layer['rank_ratio']:.3f} ({layer['compressibility']} priority)")

# Visualize singular value decay
print("\n" + "="*70)
print("SINGULAR VALUE DECAY COMPARISON")
print("="*70)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot singular values
for name, layer in [('FC', model.fc)] + [(n, c) for n, c in conv_layers[:2]]:
    if hasattr(layer, 'weight'):
        w = layer.weight.data.cpu().numpy()
        if len(w.shape) > 2:
            w = w.reshape(w.shape[0], -1)
        _, s, _ = svd(w, full_matrices=False)
        axes[0].plot(s / s[0], label=name, linewidth=2)

axes[0].set_xlabel('Index')
axes[0].set_ylabel('Normalized Singular Value')
axes[0].set_title('Singular Value Decay (Normalized)')
axes[0].set_yscale('log')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot condition numbers
cond_nums = [float(l['condition_number']) for l in layer_analysis]
layers = [l['layer'] for l in layer_analysis]
colors = ['red' if l['compressibility'] == 'High' else 'orange' if l['compressibility'] == 'Medium' else 'green'
          for l in layer_analysis]

axes[1].bar(range(len(cond_nums)), cond_nums, color=colors)
axes[1].set_xticks(range(len(layers)))
axes[1].set_xticklabels(layers, rotation=45, ha='right')
axes[1].set_ylabel('Condition Number')
axes[1].set_yscale('log')
axes[1].set_title('Condition Number by Layer')
axes[1].axhline(y=100, color='r', linestyle='--', alpha=0.5, label='High threshold')
axes[1].axhline(y=10, color='orange', linestyle='--', alpha=0.5, label='Medium threshold')
axes[1].legend()
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nAnalysis complete!")
```
