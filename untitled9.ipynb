1. COMPUTATIONAL FOUNDATIONS (Small Dataset)

1.1 normal equation

```python
import numpy as np

# set the random seed for reproducibility
np.random.seed(42)

# define the feature matrix A (4 samples, 2 features)
A = np.array([[1, 2.0],
              [1, 3.0],
              [1, 4.5],
              [1, 5.5]])

# define the target vector y (4 samples, 1 target)
y = np.array([[3.1],
              [4.9],
              [8.2],
              [10.1]])

# number of samples (m) and number of features (d)
m = len(y)
d = A.shape[1]

print("Feature matrix A:\n", A)
print("Target vector y:\n", y)
```

```python
# Normal equation: θ = (A^T A)^-1 A^T y
theta_normal = np.linalg.inv(A.T @ A) @ A.T @ y

print("Normal Equation solution θ:\n", theta_normal)
```

1.2 batch gradient descent

```python

learning_rate = 0.01
n_iterations = 1000

# initialize theta
theta_bgd = np.random.randn(d, 1)

# BGD algorithm
for iteration in range(n_iterations):
    gradients = 2/m * A.T @ (A @ theta_bgd - y)  # calculate the gradient
    theta_bgd -= learning_rate * gradients  # update theta

print("BGD solution θ:\n", theta_bgd)
```

1.3 SVD

```python
# SVD solution: θ = A+ y, where A+ is the pseudoinverse of A
U, S, VT = np.linalg.svd(A, full_matrices=False)
S_inv = np.diag(1/S)  # inverse of the singular values
A_plus = VT.T @ S_inv @ U.T  # pseudoinverse of A
theta_svd = A_plus @ y

print("SVD solution θ:\n", theta_svd)
```

1.4 numerical stability analysis

```python
# Compare θ values from the three methods

print("Comparison of θ values from three methods:")

print(f"Normal Equation θ:\n{theta_normal}")
print(f"BGD θ:\n{theta_bgd}")
print(f"SVD θ:\n{theta_svd}")

# Compute the differences between the solutions
normal_bgd_diff = np.linalg.norm(theta_normal - theta_bgd)
normal_svd_diff = np.linalg.norm(theta_normal - theta_svd)
bgd_svd_diff = np.linalg.norm(theta_bgd - theta_svd)

print(f"Difference between Normal Equation and BGD: {normal_bgd_diff}")
print(f"Difference between Normal Equation and SVD: {normal_svd_diff}")
print(f"Difference between BGD and SVD: {bgd_svd_diff}")
```

```python
# create the Collinear Matrix A_collinear

# add small noise to the second column of A to introduce multicollinearity
noise = np.random.rand(m, 1) * 0.0001
A_collinear = np.hstack((A, A[:, [1]] + noise))

print("Collinear Matrix A_collinear:\n", A_collinear)
```

```python
# Solve using Normal Equation on the Collinear Matrix

# Normal equation: θ = (A_collinear^T A_collinear)^-1 A_collinear^T y
theta_normal_collinear = np.linalg.inv(A_collinear.T @ A_collinear) @ A_collinear.T @ y

print("Normal Equation solution for Collinear Matrix θ:\n", theta_normal_collinear)
```

```python
# Solve using SVD on the Collinear Matrix

# SVD solution: θ = A+ y, where A+ is the pseudoinverse of A_collinear
U_collinear, S_collinear, VT_collinear = np.linalg.svd(A_collinear, full_matrices=False)
S_inv_collinear = np.diag(1/S_collinear)  # Inverse of the singular values
A_plus_collinear = VT_collinear.T @ S_inv_collinear @ U_collinear.T  # Pseudoinverse of A_collinear
theta_svd_collinear = A_plus_collinear @ y

print("SVD solution for Collinear Matrix θ:\n", theta_svd_collinear)
```

```python
# Numerical Analysis of Multicollinearity

# compare the results for the collinear matrix
print("Numerical Stability Analysis of Collinear Matrix:")

# differences between Normal Equation and SVD solutions for the collinear matrix
normal_svd_collinear_diff = np.linalg.norm(theta_normal_collinear - theta_svd_collinear)

print(f"Difference between Normal Equation and SVD for Collinear Matrix: {normal_svd_collinear_diff}")
```

2.1 load and clean Auto MPG

```python
import pandas as pd

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'ModelYear', 'Origin', 'CarName']

# load the dataset into a pandas DataFrame
data_mpg = pd.read_csv(url, names=column_names, na_values='?', delim_whitespace=True)

# drop rows where 'Horsepower' column has missing values
data_mpg = data_mpg.dropna(subset=['Horsepower'])

# convert 'Horsepower' column to float
data_mpg['Horsepower'] = data_mpg['Horsepower'].astype(float)

# display the first few rows of the cleaned dataset
print(data_mpg.head())
```
