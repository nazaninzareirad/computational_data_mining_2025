1. importing datasets

```python
# importing libraries
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer, load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error, r2_score, silhouette_score
from sklearn.feature_selection import SelectKBest, RFE, VarianceThreshold
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LassoCV

# Set random seed for reproducibility
SEED = 42
np.random.seed(SEED)
```

```python
# loading datasets
bc = load_breast_cancer()
X_bc, y_bc = bc.data, bc.target

iris = load_iris()
X_iris, y_iris = iris.data, iris.target

from sklearn.datasets import fetch_openml
boston = fetch_openml(name="boston", version=1)
X_boston, y_boston = boston.data, boston.target
```

2.1 correlation matrices

```python
# convert into DataFrames
df_bc = pd.DataFrame(X_bc, columns=bc.feature_names)
df_iris = pd.DataFrame(X_iris, columns=iris.feature_names)
df_boston = pd.DataFrame(X_boston, columns=boston.feature_names)

# check for missing values and basic info
print("Breast Cancer Dataset Info:")
print(df_bc.info())
print("Iris Dataset Info:")
print(df_iris.info())
print("Boston Housing Dataset Info:")
print(df_boston.info())

# correlation matrices
def plot_correlation_matrix(df, title="Correlation Matrix"):
    corr = df.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1)
    plt.title(title)
    plt.show()

plot_correlation_matrix(df_bc, "Breast Cancer - Correlation Matrix")
plot_correlation_matrix(df_iris, "Iris - Correlation Matrix")
plot_correlation_matrix(df_boston, "Boston Housing - Correlation Matrix")
```

3.1 PCA

```python
# PCA to reduce dimensions
pca_bc = PCA(n_components=0.95)  # Retain 95% variance
X_bc_pca = pca_bc.fit_transform(X_bc)

pca_iris = PCA(n_components=0.95)
X_iris_pca = pca_iris.fit_transform(X_iris)

pca_boston = PCA(n_components=0.95)
X_boston_pca = pca_boston.fit_transform(X_boston)

# plot explained variance ratio for each dataset
def plot_explained_variance(pca, title="Explained Variance"):
    explained_variance = np.cumsum(pca.explained_variance_ratio_)
    plt.plot(explained_variance, marker='o')
    plt.title(title)
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.show()

plot_explained_variance(pca_bc, "Breast Cancer PCA Explained Variance")
plot_explained_variance(pca_iris, "Iris PCA Explained Variance")
plot_explained_variance(pca_boston, "Boston Housing PCA Explained Variance")
```

5.2 SGDRegressor (Gradient-Based Optimization)

```python
# --- SGD manual training with loss tracking (works for sklearn >=0.24) ---
import numpy as np, time, matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

SEED = 42

# Use Boston data already loaded as X_boston, y_boston (ensure numpy arrays)
X = np.asarray(X_boston, dtype=float)
y = np.asarray(y_boston, dtype=float)

# Train/test split (do once, keep consistent)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)

# Standardize (fit on train only)
scaler = StandardScaler().fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s  = scaler.transform(X_test)

# Choose PCA config: either keep 95% variance or set n_components manually
pca_keep_variance = 0.95  # set to None if you want fixed components
if pca_keep_variance is not None:
    pca = PCA(n_components=pca_keep_variance, svd_solver='full', random_state=SEED)
else:
    pca = PCA(n_components=5, random_state=SEED)  # example fixed number

# Fit PCA on standardized TRAIN set and transform both train/test
X_train_pca = pca.fit_transform(X_train_s)
X_test_pca  = pca.transform(X_test_s)

print("Original dim:", X_train_s.shape[1], "PCA dim:", X_train_pca.shape[1])

# Function: train SGDRegressor via partial_fit and track training loss per epoch
def train_sgd_with_loss(X_tr, y_tr, X_val, y_val,
                        max_epochs=1000, tol=1e-6,
                        learning_rate='invscaling', eta0=0.01, random_state=SEED):
    # Initialize regressor with warm_start not needed when using partial_fit
    sgd = SGDRegressor(max_iter=1, tol=None, learning_rate=learning_rate, eta0=eta0,
                       random_state=random_state, shuffle=False)  # we will shuffle manually
    n_samples = X_tr.shape[0]
    # For partial_fit we must call it once with classes for classification; for regression it's fine directly
    train_losses = []
    val_losses = []
    prev_loss = np.inf
    epoch = 0
    idx = np.arange(n_samples)
    while epoch < max_epochs:
        # shuffle each epoch for SGD
        np.random.shuffle(idx)
        X_sh = X_tr[idx]
        y_sh = y_tr[idx]
        # We'll do one full pass over data with partial_fit in minibatches (here: single batch)
        # You can split into mini-batches if you want; here we do full-batch partial_fit which emulates epochs.
        sgd.partial_fit(X_sh, y_sh)
        # compute training & validation loss (MSE)
        y_pred_train = sgd.predict(X_tr)
        y_pred_val = sgd.predict(X_val)
        train_mse = mean_squared_error(y_tr, y_pred_train)
        val_mse = mean_squared_error(y_val, y_pred_val)
        train_losses.append(train_mse)
        val_losses.append(val_mse)
        epoch += 1
        # check tol convergence based on training loss change
        if abs(prev_loss - train_mse) < tol:
            break
        prev_loss = train_mse
    # attach some useful attrs
    sgd._manual_n_iter = epoch
    sgd._train_losses = np.array(train_losses)
    sgd._val_losses = np.array(val_losses)
    return sgd

# Train on RAW standardized data
sgd_raw = train_sgd_with_loss(X_train_s, y_train, X_test_s, y_test,
                              max_epochs=500, tol=1e-6, eta0=0.01)
y_pred_raw = sgd_raw.predict(X_test_s)
print("SGDRegressor on Raw Data:")
print("R:", r2_score(y_test, y_pred_raw))
print("MSE:", mean_squared_error(y_test, y_pred_raw))
print("Iterations (manual):", sgd_raw._manual_n_iter)

# Train on PCA data
sgd_pca = train_sgd_with_loss(X_train_pca, y_train, X_test_pca, y_test,
                              max_epochs=500, tol=1e-6, eta0=0.01)
y_pred_pca = sgd_pca.predict(X_test_pca)
print("\nSGDRegressor on PCA Data:")
print("R:", r2_score(y_test, y_pred_pca))
print("MSE:", mean_squared_error(y_test, y_pred_pca))
print("Iterations (manual):", sgd_pca._manual_n_iter)

# Plot training loss curves (MSE) for comparison
plt.figure(figsize=(9,6))
plt.plot(sgd_raw._train_losses, label=f'Raw (dim={X_train_s.shape[1]})')
plt.plot(sgd_pca._train_losses,  label=f'PCA (dim={X_train_pca.shape[1]})')
plt.yscale('log')  # log scale often clearer for convergence
plt.xlabel('Epoch')
plt.ylabel('Train MSE (log scale)')
plt.title('SGDRegressor training loss (manual)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import time

# train KMeans on raw data
kmeans_boston_raw = KMeans(n_clusters=3, random_state=42)
start_time = time.time()
kmeans_boston_raw.fit(X_boston)
end_time = time.time()
inertia_raw = kmeans_boston_raw.inertia_
silhouette_raw = silhouette_score(X_boston, kmeans_boston_raw.labels_)
n_iter_raw = kmeans_boston_raw.n_iter_

# train KMeans on PCA data
kmeans_boston_pca = KMeans(n_clusters=3, random_state=42)
start_time_pca = time.time()
kmeans_boston_pca.fit(X_boston_pca)
end_time_pca = time.time()
inertia_pca = kmeans_boston_pca.inertia_
silhouette_pca = silhouette_score(X_boston_pca, kmeans_boston_pca.labels_)
n_iter_pca = kmeans_boston_pca.n_iter_

# results
print("KMeans on Raw Data:")
print(f"Inertia: {inertia_raw}, Silhouette: {silhouette_raw}, Time: {end_time - start_time}s, Iterations: {n_iter_raw}")

print("KMeans on PCA Data:")
print(f"Inertia: {inertia_pca}, Silhouette: {silhouette_pca}, Time: {end_time_pca - start_time_pca}s, Iterations: {n_iter_pca}")
```
